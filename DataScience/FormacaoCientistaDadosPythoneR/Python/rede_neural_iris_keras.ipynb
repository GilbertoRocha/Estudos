```
from sklearn import datasets
from sklearn.model_selection import train_test_split
from yellowbrick.classifier import ConfusionMatrix
from sklearn.metrics import confusion_matrix
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
```

```
# carrega as info
base = datasets.load_iris()
previsores = base.data
classe = base.target
```

```
# cria as classes dummy para a saida.
# como tem 3 classes, serao necessarios 3 neuronios de saida, então é criado uma coluna para cada neuronio
classe_dummy = np_utils.to_categorical(classe)
print(classe_dummy)
```

```
# separa os dados de teste
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(previsores, classe_dummy, test_size = 0.3, random_state = 0)
```

```
# cria o modelo com 5 neuronios
modelo = Sequential()
# camada de entrada
modelo.add(Dense(units= 5, input_dim = 4))
# camada escondida
modelo.add(Dense(units = 4))
# camada saida usando a softmax, sendo indicada para classificação com mais de 2 classes
modelo.add(Dense(units = 3, activation = 'softmax'))

modelo.summary()
```

```
# parametriz o modelo
# loss = categorical_crossentropy, calculo de custo indicado para 2 ou mais clases
modelo.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

modelo.fit(X_treinamento, y_treinamento, epochs = 1000, validation_data = (X_teste, y_teste))
```

```
previsoes = modelo.predict(X_teste)
# traz o percentual de cada classe
print(previsoes)
```

```
# atribui a classe que tiver mais de 50% de chance
previsoes = (previsoes > 0.5)
print(previsoes)
```

```
# verifica a precisao
#pega somente o index do atributo (indicando a classe)
y_teste_matrix = [np.argmax(t) for t in y_teste]
y_previsao_matrix = [np.argmax(t) for t in previsoes]

confusao = confusion_matrix(y_teste_matrix, y_previsao_matrix)
print(confusao)
```
